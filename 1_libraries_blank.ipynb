````python
# Load CSV using Pandas from URL
import pandas as pd
from IPython.display import display, HTML
data = pd.read_csv('guia_fasecolda.csv')

print(data.columns)

#Realiza diferentes filtros sobre los datos y describelos

# 1.Filtrar vehículos con potencia mayor a 400 HP (columna Potencia):
filter = data[data['Potencia'] > 400]
#print(filter)

# 2.Filtrar vehículos con potencia mayor a 400 HP (columna Potencia) y cilindraje mayor a 6000
filter = filter[filter['Cilindraje'] > 6000]
#print(filter)

# 3.Filtrar vehículos con potencia mayor a 400 HP (columna Potencia), cilindraje mayor a 6000 y combustible sea igual a Gasolina
filter = filter[filter['Combustible'] == 'GSL']
#print(filter)

# Integración de tres funciones de pandas para describir los datos filtrados.
print('Descripción:')
print(filter.describe())

print('Información:')
print(filter.info())

filter_corr = filter[['Potencia', 'Cilindraje']]
print('Correlación:')
print(filter_corr.corr())


#display(HTML((data.head().to_html())))
#print(df[data['2015']>50000])
````

````python

%matplotlib inline
import pandas as pd # Importamos la libreria de Pandas
import matplotlib.pyplot as plt
import seaborn as sns  # Seaborn para mejorar la visualización

plt.figure()
data[['Cilindraje','Peso']].plot(kind = 'box',ylim=(-1000,10000))
plt.title('mi primera grafica')
plt.show()

# Filtro: Vehículos con potencia mayor a 400 HP, cilindraje mayor a 6000 y combustible = Gasolina
filter_data = data[(data['Potencia'] > 400) & (data['Cilindraje'] > 6000) & (data['Combustible'] == 'GSL')]

## 1. Gráfico de Dispersión: Potencia vs Cilindraje
plt.figure(figsize=(8,5))
sns.scatterplot(x=filter_data['Cilindraje'], y=filter_data['Potencia'], color='b', alpha=0.6)
plt.xlabel("Cilindraje (cc)")
plt.ylabel("Potencia (HP)")
plt.title("Relación entre Cilindraje y Potencia")
plt.grid(True)
plt.show()

## 2. Histograma con Curva de Distribución (Campana de Gauss)
plt.figure(figsize=(8,5))
sns.histplot(filter_data['Potencia'], bins=10, kde=True, color='g') # kde=True para la curva de Gauss
plt.xlabel("Potencia (HP)")
plt.ylabel("Frecuencia")
plt.title("Distribución de Potencia")
plt.grid(True)
plt.show()

````
